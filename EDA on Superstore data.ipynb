**Visualization**

```python
#sns.countplot(x=df['Sub-Category'])
y=['Ship Mode','Segment','Region','Category']
for i in y:
    sns.catplot(x=i,kind='count',data=df)
```

```python
plt.figure(figsize=(15,5))
df['State'].value_counts().plot(kind='bar')
plt.xlabel('State',fontsize=15)
plt.xticks(fontsize=12)
plt.ylabel("Count",fontsize=15)
plt.yticks(fontsize=12)
```

```python
plt.figure(figsize=(8,5))
df['Sub-Category'].value_counts().plot(kind='bar')
plt.xlabel('Sub-Category',fontsize=15)
plt.xticks(fontsize=12)
plt.ylabel("Count",fontsize=15)
plt.yticks(fontsize=12)
```

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot(kind='pie')
```

**Binders are being sold in large quantity**

```python
grouped=df.groupby(['Sub-Category'])
g1=grouped['Profit'].agg(np.sum)
print(g1)
```

```python
plt.figure(figsize=(10,7))
g1.plot(kind='bar',color='teal')
plt.xlabel('Sub-Category',fontsize=20)
plt.xticks(fontsize=15)
plt.ylabel("Count",fontsize=20)
plt.yticks(fontsize=15)
```

**We can see that overall sales of Copiers is generating maximum profit and Bookcases, Supplies and Tables are in loss.**

```python
plt.figure(figsize=(30,15))
sns.countplot(x='Sub-Category',hue='Region',data=df)
plt.legend(fontsize=30)
plt.xlabel('Sub-Category',fontsize=30)
plt.xticks(rotation=90,fontsize=25)
plt.ylabel("Count",fontsize=30)
plt.yticks(fontsize=25)
```

```python
grouped=df.groupby(['Sub-Category','Region'])
g2=grouped['Profit'].agg(np.sum)
g2=pd.DataFrame(g2)
print(g2)
```
